```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
print(train.shape)
print(test.shape)
```

```python
print(train.columns)
print(test.columns)
```

```python
full = train.append(test,ignore_index = True,sort = True)
print(full.head())
```

```python
print(full.shape)
full.info()
full.describe()
```

```python
full['Age'] = full['Age'].fillna(full['Age'].mean())
full['Fare'] = full['Fare'].fillna(full['Fare'].mean())
full.describe()
```

```python
full['Embarked'].value_counts()
```

```python
full['Cabin'] = full['Cabin'].fillna('U')
full['Embarked'] = full['Embarked'].fillna('S')
full.info()
```

```python
'''
Categorical data: Cabin, Embarked, Pclass, Sex, Parch, Sibsp
string data: Name, Ticket
Series data: Age, Fare, PassengeId
'''
```

```python
sex_Dict = {'female':0, 'male':1}
full['Sex'] = full['Sex'].map(sex_Dict)
full['Sex'].head()
```

```python
EmbarkedDF = pd.DataFrame()
EmbarkedDF = pd.get_dummies(full['Embarked'],prefix = 'Embarked')
EmbarkedDF.head()
```

```python
full = pd.concat([full, EmbarkedDF],axis = 1)
full.drop('Embarked',axis = 1,inplace = True)
full.head()
```

```python
full['Pclass'].head()
```

```python
PclassDF = pd.DataFrame()
PclassDF = pd.get_dummies(full['Pclass'], prefix='Pclass')
PclassDF.head()
```

```python
full = pd.concat([full, PcalssDF],axis = 1)
full.drop('Pclass',axis = 1,inplace = True)
full.head()
```

```python
CabinDF = pd.DataFrame()
CabinDF = full['Cabin'].map(lambda a: a[0])
CabinDF.head() 
```

```python
CabinDF = pd.get_dummies(CabinDF ,prefix = 'Cabin')
CabinDF.head()
```

```python
full = pd.concat([full, CabinDF],axis = 1)
full.drop('Cabin',axis = 1,inplace = True)
full.head()
```

```python
FamilyDF = pd.DataFrame()
FamilyDF['familySize'] = full['Parch'] + full['SibSp'] + 1
FamilyDF.head()
FamilyDF['familySize'].value_counts()
```

```python
FamilyDF['familySingle'] = FamilyDF['familySize'].map(lambda b:1 if b<=1 else 0)
FamilyDF['familyMiddle'] = FamilyDF['familySize'].map(lambda b:1 if 2<=b<=4 else 0)
FamilyDF['familyLarge'] = FamilyDF['familySize'].map(lambda b:1 if b>=5 else 0)
FamilyDF.head()
```

```python
full = pd.concat([full, FamilyDF], axis = 1)
full.head()
```

```python
full['Name'].head()
```

```python
'''
name1 = 'Braund, Mr. Owen Harris'
str1 = name.split(',')[1]
str2 =str1.split('.')[0]
str3 = str2.strip()

'''
```

```python
def firstget(name):
    str1 = name.split(',')[1]
    str2 =str1.split('.')[0]
    str3 = str2.strip()     # delete space
    return str3

titleDF = pd.DataFrame()
titleDF['Title'] = full['Name'].map(firstget)
titleDF.head()
```

```python
titleDF['Title'].value_counts()
```

```python
title_mapDict = {
    'Mr':'Mr',
    'Miss':'Miss',
    'Mrs':'Mrs',            
    'Master':'Master',           
    'Rev':'Officer',              
    'Dr':'Officer',                 
    'Col':'Officer',                
    'Ms':'Mrs',                
    'Major':'Officer',              
    'Mlle':'Miss',              
    'Sir':'Royalty',   
    'Dona':'Royalty',               
    'Jonkheer':'Royalty' ,          
    'Capt':'Officer',              
    'Don':'Royalty',                
    'the Countess':'Royalty',      
    'Mme':'Mrs',               
    'Lady':'Royalty'               
}
titleDF['Title'] = titleDF['Title'].map(title_mapDict)
titleDF.head()
```

```python
titleDF = pd.get_dummies(titleDF)
titleDF.head()
```

```python
full = pd.concat([full,titleDF], axis =1)
full.head()
```

```python
full.info()
```

```python
corrDF = full.corr()
corrDF
```

```python
corrDF['Survived'].sort_values(ascending = False)
```

```python
full_x = pd.concat([titleDF,PclassDF,full['Fare'],CabinDF,EmbarkedDF,FamilyDF,full['Age'],full['Sex']],axis=1)
full_x.head()
```

```python
sourceRow = 891
```

```python
source_x = full_x.loc[0:sourceRow -1,:]
source_y = full.loc[0:sourceRow -1,'Survived']
pre_x = full_x.loc[sourceRow:,:]

print(source_x.shape)
print(source_y.shape)
print(pre_x.shape)
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(source_x,source_y,test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_x,train_y)
model.score(test_x,test_y)
```

```python
pre_y = model.predict(pre_x)
pre_y = pre_y.astype('int')
pre_y
```

```python
Passenger_ID = full.loc[sourceRow:,'PassengerId']
preDF = pd.DataFrame({'PassengerId':Passenger_ID, 'Survived':pre_y})
preDF.head()
```

```python
preDF.to_csv('titanic_pred2.csv',index=False)
```
